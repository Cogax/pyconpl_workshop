# Amazon Lambda

AWS Lambda takes some code and executes it and you don't have to worry about managing the infrastructure.

```python
# An example of a Lambda function in Python
def handler_name(event, context):
    return {"message": "Hello"}
```

## No worries about managing infrastructure
* No web servers
* No web framework
* Just pure Python

![infrastructure](https://cdn.meme.am/instances/63594787.jpg)

# Pros of AWS Lambda

* You really don't have to worry about the infrastructure
* Could be very useful if you can think in a "lambda way"
* Versioning your lambda functions
* You are paying for something only when you use it
* Good integration with all the AWS Services

# Cons of AWS Lambda

* You need to learn a new paradigm
* You have to worry about the deployment packages
* The deployment package could be only 50MB
* The lambda function can run for no more than 300 secs
* Ephemeral disk capacity 512MB
* Lack of documentation, best practies and still Python 2.7

# Prices

You get a bill for the number of requestes and for the duration of the executions.

The first 1.000.000 requests and 400,000 GB-seconds of execution are for free.

https://aws.amazon.com/lambda/pricing/
